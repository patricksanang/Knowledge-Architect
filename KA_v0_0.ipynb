```python
#function to get sentences from a paragraph
def get_sentences_from_qa_answers(paragraph):
  return sent_tokenize(paragraph)


#Q/A embedder
def embedder(QAs, model):
  output_QAs = []
  embeddings_question = []
  embeddings_answer = []
  tokens = []
  sentences = []

  if len(QAs) == 0 :
    return output_QAs

  # Encode all sentences
  for x in range(len(QAs)):
    tokens = sent_tokenize(QAs[x]["answer"], language="french")
    embeddings_question.append(model.encode(QAs[x]["question"]))
    sentences.append(tokens)
    embeddings_answer.append(model.encode(tokens))

  #print(sentences[0])
  #print("----separateur-----")
  #print(sentences[1])
  return zip(QAs, sentences, embeddings_question, embeddings_answer)

def top_k_mostsim(k, cos_sim):

  #we get the top-k elements of the flattened version of our tensor

  tops_k, indexes = torch.topk(cos_sim.flatten(), k, sorted=True)
  indexes = np.array(np.unravel_index(indexes.numpy(), cos_sim.shape)).T

  return zip(tops_k, indexes)
```
